```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_excel('NASA_Labs_Facilities.xlsx', header=1, index_col=0)
data.head(5)
```

```python
data.rename(columns = {'Unnamed: 1':'Government Department'}, inplace = True) 
data.head(5)
```

```python
print(data.shape)
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data['Center'].unique()
```

```python
print(data['Center'].value_counts())
plt.figure(figsize=(20,6))
sns.countplot(y=data['Center'], data=data, order = data['Center'].value_counts().index)
```

```python
plt.figure(figsize=(20,6))
sns.countplot(y=data['Status'], data=data, order = data['Status'].value_counts().index)
print(data['Status'].value_counts())
```

```python
print(data['State'].value_counts())
plt.figure(figsize=(20,6))
plt.title("Which state has most clinical trials of MSC")
sns.countplot(x="State", data=data, order = data['State'].value_counts().iloc[:10].index)
plt.xlabel('State')
plt.ylabel('Count')
```
